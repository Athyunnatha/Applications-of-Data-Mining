```python
# Supress Warnings

import warnings
warnings.filterwarnings('ignore')

# Import the numpy and pandas package

import numpy as np
import pandas as pd

# Data Visualisation
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
advertising = pd.DataFrame(pd.read_csv('/content/SLR_Salary_dataset.csv'))
advertising.head()
```

```python
# Checking Null values
advertising.isnull().sum()*100/advertising.shape[0]
# There are no NULL values in the dataset, hence it is clean.
```

```python
# Outlier Analysis
fig, axs = plt.subplots(3, figsize = (5,5))
plt1 = sns.boxplot(x=advertising['YearsExperience'], ax = axs[0]) # Changed 'TV' to 'YearsExperience'
plt2 = sns.boxplot(x=advertising['Salary'], ax = axs[1]) # Changed 'Newspaper' to '
```

```python
sns.boxplot(advertising['Salary'])
plt.show()
```

```python
# Let's see how Salary are related with YearsExperience using scatter plot.
sns.pairplot(advertising, x_vars=['YearsExperience'], y_vars='Salary', height=4, aspect=1, kind='scatter')
plt.show()
```

```python
# Let's see the correlation between different variables.
sns.heatmap(advertising.corr(), cmap="YlGnBu", annot = True)
plt.show()
```
